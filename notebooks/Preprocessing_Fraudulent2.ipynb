```python
import pandas as pd

file_path = '../data/raw/Fraudulent2.csv'
df = pd.read_csv(file_path)
print(df.head())
print(df.info())
```

```python
df.isnull().sum()
```

```python
df.fillna({"kyc_verified": "No"}, inplace=True)
df.dropna(subset=["transaction_amount"], inplace=True)
```

```python
df.drop_duplicates(subset=['transaction_id'], keep='first', inplace=True)
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['channel'] = df['channel'].str.title()
df['transaction_amount'] = df['transaction_amount'].astype(float)
```

```python
df.head()
```

```python
df.info()
```

```python
df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.day_name()
df['is_high_value'] = (df['transaction_amount'] > 50000).astype(int)
```

```python
df.head()
```

```python
df['is_high_value'].value_counts()
```

```python
df.to_csv('../data/processed/Fraudulent2_processed.csv', index=False)
```
